```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("./SORL-Deep.csv")
df2 = pd.read_csv("./SORL-Tabular.csv")
df3 = pd.read_csv("./MORL-Tabular.csv")
df4 = pd.read_csv("./MORL-Deep.csv")
df5 = pd.read_csv('./Uniform-Random.csv')
```

```python
plt.plot(df1["Reward"],label="SORL-Deep",c="blue")
plt.plot(df2["Reward"],label="SORL-Tabular",c="red")
plt.plot(df3["Reward"],label="MORL-Tabular",c="green")
plt.plot(df4['Reward'],label="MORL-Deep",c='orange')
plt.plot(df5['Reward'],label="Uniform Random Policy",c="purple")
plt.legend()
plt.xlabel("Timestep")
plt.ylabel("Cummulative Reward")
plt.title("Cummulative Reward")
# fig,axes = plt.subplots(1,2,figsize=(15,3))
# axes[0].plot(df1["Reward"],label="SORL",c="b")
# axes[1].plot(df2["Reward"],label="All offload",c="r")
# axes[2].plot(df3["Reward"],label="All Local",c="k")
# axes[3].plot(df4['Reward'],label="Uniform Random",c='y')
```

```python
plt.plot(df1["Running Latency"],label="SORL-Deep",c="blue")
plt.plot(df2["Running Latency"],label="SORL-Tabular",c="red")
plt.plot(df3["Running Latency"],label="MORL-Tabular",c="green")
plt.plot(df4["Running Latency"],label="MORL-Deep ",c="orange")
plt.plot(df5['Running Latency'],label="Uniform Random Policy",c="purple")
plt.legend()
plt.xlabel("Timestep")
plt.ylabel("Running Latency (in timesteps)")
plt.title("Running Latency")
```
